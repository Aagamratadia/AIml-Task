```python
import numpy as np
import matplotlib.pyplot as plt

frequency = 1000
time = np.linspace(0, 2e-3, int(2e-3 / 10e-6))
signal = np.sin(2 * np.pi * frequency * time)

plt.figure(figsize=(10, 6))
plt.plot(time, signal, color='green')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Continuous Signal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

ts = np.arange(0, 2e-3, 0.1e-3)
c = np.sin(2 * np.pi * frequency * ts)

plt.figure(figsize=(10, 6))
plt.stem(ts, c, 'k')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Discrete Signal')
plt.grid(True)
plt.show()
```
