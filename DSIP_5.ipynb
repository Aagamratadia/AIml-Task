```python
import numpy as np


def dft_step_by_step(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)

    for k in range(N):
        print(f"\nCalculating X({k}):")
        X_k = 0

        for n in range(N):
            exp_term = np.exp(-2j * np.pi * k * n / N)

            X_k += x[n] * exp_term

            print(f"x({n}) = {x[n]} * exp(-j * 2π * {k} * {n} / {N}) = {x[n]} * {exp_term} = {x[n] * exp_term}")

        X[k] = X_k
        print(f"X({k}) = {X_k}")

    return X

x_4 = np.array([1, 2, 3, 4])

X_4_step_by_step = dft_step_by_step(x_4)

print("\nFinal 4-point DFT Result:")
print(X_4_step_by_step)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)

    for k in range(N):
        X_k = 0

        for n in range(N):
            exp_term = np.exp(-2j * np.pi * k * n / N)

            X_k += x[n] * exp_term

        X[k] = X_k

    return X


x_4 = np.array([1, 2, 3, 4])
X_4 = dft(x_4)
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.stem(np.arange(len(x_4)), x_4, use_line_collection=True)
plt.title('Input Sequence (Time Domain)')
plt.xlabel('n')
plt.ylabel('x(n)')

plt.subplot(2, 2, 3)
plt.stem(np.arange(len(X_4)), np.abs(X_4), use_line_collection=True)
plt.title('Magnitude of DFT (Frequency Domain)')
plt.xlabel('k')
plt.ylabel('|X(k)|')

plt.subplot(2, 2, 4)
plt.stem(np.arange(len(X_4)), np.angle(X_4), use_line_collection=True)
plt.title('Phase of DFT (Frequency Domain)')
plt.xlabel('k')
plt.ylabel('Phase of X(k)')

plt.tight_layout()
plt.show()
```
